```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
RgDFNTraj = pd.read_csv("alpha_S/analysis_scripts/rg_dssp/Rad.dat", delim_whitespace=True, header=None, names=["Sim1"])#,"Sim5","Sim6", "Sim7", "Sim8","Sim9","Sim10", "Sim11", "Sim12", "Sim13", "Sim14", "Sim15", "Sim16" ])
bins = np.arange(7.5, 80, 0.5) #define bins as 15 to 45
bin_edges = 0.5*(bins[1:]+bins[:-1])
RgDFNTraj_Hist = pd.DataFrame(columns=RgDFNTraj.columns, index=bin_edges).fillna(0.0)

for sim in RgDFNTraj.columns:
    hist, edges = np.histogram(RgDFNTraj[sim].values, bins=bins, density=True)
    RgDFNTraj_Hist[sim] = hist
    
cpalette = sns.color_palette(palette="hls", n_colors=RgDFNTraj_Hist.shape[1]+2)
figRgHist = plt.figure(131512, figsize=(6,4))
axRgH = figRgHist.add_subplot(111)

axRgH = RgDFNTraj_Hist.plot(color=cpalette, ax=axRgH, linewidth=2, legend=False) ## Plot histograms
axRgH.set_ylim(0.0,0.08)
RgHandles, RgLabels = axRgH.get_legend_handles_labels()

#axRgH.set_title('HOX-SCR',fontsize=16, weight='bold')
#axRgH.legend(RgHandles, RgLabels, frameon=False,prop={'size':11})
#axRgH.set_ylabel('Frequency', fontsize=16, weight='bold')
axRgH.set_xlabel(r'$\mathbf{R_{g} \ \ (\AA)}$', fontsize=24)
axRgH.tick_params(labelsize=24)

axRgH.text(-9,0.079,'H', weight='bold',fontsize=24)

#axRgH.vlines(x=RgDFNTraj['Sim1'].mean(), ymin = 0, ymax = 0.08, linewidth=2, color ="r", linestyle ="--")

axRgH.axvspan(RgDFNTraj['Sim1'].mean()-6.0, RgDFNTraj['Sim1'].mean()+6.0, 
        alpha=0.25, color='red')

axRgH.vlines(x=(2.54 * (140**0.522)), ymin = 0, ymax = 0.08, linewidth=2, color ="black")

plt.savefig("Rg_as.png",bbox_inches='tight', dpi=300)
```
